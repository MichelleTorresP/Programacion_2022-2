```python
import json
import requests
from pprint import pprint

response = requests.get(
"https://fp.josedomingo.org/lmgs/u05/ej2.json")
web = json.loads(response.text)

#imprime el json cargado
pprint(web)
```

```python
titulo=[] #Punto N°1
horas=[] #Punto N°2
urls=[] #Punto N°3
x={} #Punto N°5
for curso in web:
  for prof in curso["Profesorado"]:
    profe = prof["NombreCompleto"]
    if a in x:
      x[a]+=", "
      x[a]+=profe
    else:
      x[a]=profe
  a=curso["Titulo"]
  url = curso["URL"]
  titulo.append(a)

  if curso["Horas"] >2:
    horas.append(a)

  if curso["TipoFormacion"] == "NoPresencial" or "No Presencial":
    urls.append(url)


print(f"1. Hay {len(titulo)} pruebas de idiomas descritas en el documento")
for h in horas:
  print(f"2. {h}")
for u in urls:
  print("3.", u)
```
